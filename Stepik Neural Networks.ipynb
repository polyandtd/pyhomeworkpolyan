Найдите точку локального максимума функции $\ f(x, y) = x^3 + 4y^3 - 3x - 3y $ .

Запишите ответ в виде двух чисел (координата по x, координата по y) в десятичном формате через запятую, отделяя целую часть от дробной точкой (например, $ 2.5$, $-3 $, если вы считаете, что $\max\left\lbrace f(x,y) \middle\vert x,y\in\mathbb R\right\rbrace = f(2.5, -3)$.

```python
x = symbols('x')
expr = x**3 + 4* y**3 - 3*x - 3*y
```

```python
expr.diff(x)
```

_Нашли частную производную по x. Если приравнять к 0, то x будет равен $+-1 $_

```python
expr.diff(y)
```

_Нашли частную производную по y. Если приравнять к 0, то y будет равен $+-0.5 $_

У нас получилось 4 точки : $ (1, 0.5) $, $(-1, 0.5)$, $(1, -0.5)$, $(-1, -0.5)$. Теперь найдем значения, при которых функция принимает максимальное

```python
expr.replace(x, 1).replace(y, 0.5), expr.replace(x, -1).replace(y, 0.5), expr.replace(x, 1).replace(y, -0.5), expr.replace(x, -1).replace(y, -0.5)
```
